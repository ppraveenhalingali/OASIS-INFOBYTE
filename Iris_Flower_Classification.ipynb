```python
import pandas as pd

# Load the dataset (make sure Iris.csv is uploaded to your Colab environment)
df = pd.read_csv('Iris.csv')

print("--- DataFrame Head (First 5 Rows) ---")
print(df.head())

print("\n--- Column Information and Data Types ---")
print(df.info())
```

```python
from sklearn.preprocessing import LabelEncoder

# Drop the unnecessary 'Id' column (axis=1 means drop a column)
df = df.drop('Id', axis=1)

# Initialize the LabelEncoder
le = LabelEncoder()

# Fit and transform the 'Species' column to get numerical labels (0, 1, 2)
df['Species_Encoded'] = le.fit_transform(df['Species'])

# Print the mapping to know which number corresponds to which species
print("--- Species to Numerical Mapping ---")
for i, species in enumerate(le.classes_):
    print(f"{i}: {species}")

print("\n--- DataFrame Head after Encoding ---")
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

# Define Features (X) and Target (y)
# X contains the four measurement columns
X = df.drop(['Species', 'Species_Encoded'], axis=1)

# y contains the new numerical target labels
y = df['Species_Encoded']

# Split the data: 70% for training and 30% for testing
# random_state=42 ensures you get the same split every time
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"Training set size (70%): {len(X_train)} samples")
print(f"Testing set size (30%): {len(X_test)} samples")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the KNN model. We choose k=5 neighbors.
model = KNeighborsClassifier(n_neighbors=5)

# Train the model (this is where the learning happens)
model.fit(X_train, y_train)

print("K-Nearest Neighbors Model Training Complete!")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Use the trained model to predict the species for the test set
y_pred = model.predict(X_test)

# Calculate the Accuracy (percentage of correct predictions)
accuracy = accuracy_score(y_test, y_pred)

# Generate the Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"--- Model Evaluation Results (KNN) ---")
print(f"Accuracy Score: {accuracy*100:.2f}%")
print("\nConfusion Matrix (Rows=Actual, Columns=Predicted):")
print("-----------------------------------------")
print(conf_matrix)
```
